# Task 2: Create age groups

````python
bins = [0, 2, 18, 35, 65, 105]
names = ['<2', '2-18', '18-35', '35-65', '65+']

df['AgeGroup'] = pd.cut(df['Age'], bins, labels=names)

print(df.AgeGroup)
print(df.head())
````

# Task 3: Filter the female candidate data 

````python
data= df[df.Gender == 'Female']
data
````

# Task 4: Convert Student status from Yes/No to 0/1

````python
df['Student'] = df['Student'].map({'Yes': 1, 'No': 0})
df.head()
````
